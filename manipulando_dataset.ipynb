```python
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

fullPatient = pd.read_csv('https://raw.githubusercontent.com/laderast/cvdNight1/master/data/fullPatientData.csv', sep=',', encoding='utf-8')

food = pd.read_csv('https://raw.githubusercontent.com/bonamicojedi/esp32/main/exported_data.csv', sep=',', encoding='utf-8')
```

# **Dimensões do Banco de dados:**

```python
print('Full Patient: ', fullPatient.shape)
print('Food: ', food.shape)
```

# **20 Primeiras Linhas do Banco de Dados**

```python
print('BD Full Patient: \n')
print(fullPatient.head(20))
print('\nBD FOOD: \n')
print(food.head(20))
```

# **Resumo dos Atributos:**

```python
print('BD Full Patient:\n',fullPatient.describe())
print('\n Bd Food: \n',food.describe())
```

# **Agrupando os dados por um dos atributos**

```python
print('BD Full Patient: \n',fullPatient.groupby('bmi').size())

print('\nBD Food: \n', food.groupby('Protein (g)').size())
```
